# Linear and Logistic Regression
In this section, we'll be working with `pystatsmodels` to perform linear/logistic regression

```python
import numpy as np
import pandas as pd

CEOcomp = pd.read_csv("CEOcomp.csv")
CEOcomp
```

## Linear Regression

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
CEO_linReg = smf.ols(formula="TotalCompensation ~ Years + ChangeStockPrice + ChangeCompanySales + MBA", data = CEOcomp)
res = CEO_linReg.fit()
res.summary()
```

Quantities of interest can be extracted directly from the fitted model. Type dir(results) for a full list. Here are some examples:

```python
res.params
```

```python
print('Standard errors')
res.bse
```

```python
res.predict()
```

```python
res.conf_int()
```

## Splitting data into training and test sets

```python
titanic_passengers = pd.read_csv("TitanicPassengers.csv").fillna(0)
titanic_passengers
```

```python
indices = np.arange(len(titanic_passengers))
np.random.shuffle(indices)
len(indices), indices
```

```python
data_train, data_test = titanic_passengers.ix[indices[:5*756/6],:], titanic_passengers.ix[indices[5*756/6:],:]
```

```python
len(data_train), data_train
```

```python
len(data_test), data_test
```

## Logistic Regression

```python
# not always stable!
titanic_logReg = smf.glm(formula = "Survived ~ Class + Age + Sex", 
                         data = data_train, 
                         family = sm.families.Binomial()).fit()
titanic_logReg.summary()
```

```python
train_predicted = np.round(titanic_logReg.predict()) #Compute predicted probabilities on training data
pd.crosstab(data_train.Survived, train_predicted, rownames=['Survived'], colnames=['Prediction'])
```

```python
test_predicted = np.round(titanic_logReg.predict(exog = data_test)) #Compute predicted probabilities on training data
pd.crosstab(data_test.Survived, test_predicted, rownames=['Survived'], colnames=['Prediction'])
```

Compute percentage correct (overall accuracy)
