# Hierarchical and K-Means Clustering

## Hierarchical Clustering
We'll first cover hierarchical clustering

```python
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster

from sklearn import datasets
iris = datasets.load_iris()
```

```python
distanceMatrix = pdist(iris.data, 'euclidean') # other options: 'minkowski', 'cityblock' (manhattan), etc
distanceMatrix
```

```python
# Valid methods when the raw observations are omitted are 'single', 'complete', 'weighted', and 'average'
fig, ax = plt.subplots(figsize=(15,8))
dendrogram(linkage(distanceMatrix, method='weighted'));
```

This diagram will help us decide how many clusters are appropriate for this problem. The height of the vertical lines represents the distance between the points that were combined into clusters. The record numbers are listed among the bottom (usually hard to see). The taller the lines, the more likely it is that clusters should be separate. Two or three clusters would be appropriate here.

```python
fig, ax = plt.subplots(figsize=(18,6))
d = dendrogram(linkage(distanceMatrix, method='weighted'),color_threshold=2);
```

```python
fcluster(linkage(distanceMatrix, method='weighted'),2,'distance') # provides the labels for each terminal node/leaf
```

```python
pd.crosstab(iris.target,
            fcluster(linkage(distanceMatrix, method='weighted'),2,'distance'))
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['hgroups'] = fcluster(linkage(distanceMatrix, method='weighted'),2,'distance')
iris_df
```

The means of each of the attributes will give us the centroids of the clusters.

```python
iris_df.groupby('hgroups').mean()
```

## K-Means clustering
K-means clustering requires that we have an initial guess as to how many clusters there are.  We will initialize it to 3 in this case, but if we didn't know, we could always try multiple values and experiment

Run a k-means cluster with 3 clusters and 100 iterations (centroids recomputed and points reassigned each time)

```python
iris_KMC.fit(iris.data)
```

# Support Vector Machines
Build SVM model for iris data set (since SVM is easier to visualize with smaller datasets with continuous attributes)

First, we want to subset the dataset to only keep two attributes (so we can easily visualize the model)

```python
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data[:, :2] # take only the first two features
Y = iris.target
X
```

```python
Y
```

You can find the description of the full dataset here: [The Iris Dataset](http://scikit-learn.org/stable/auto_examples/datasets/plot_iris_dataset.html)

```python
from sklearn import svm

h = .02  # step size in the mesh

# we create an instance of SVM and fit out data. We do not scale our
# data since we want to plot the support vectors
C = 1.0  # SVM regularization parameter
svc = svm.SVC(kernel='linear', C=C).fit(X, Y)

# gamma controls how well the model will fit the data. Larger gamma will fit the data more exactly.
# Try gamma = 100 and gamma = 0.1
rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(X, Y)

# degree = degree of polynomial used. Different values will often give very different results.
poly_svc = svm.SVC(kernel='poly', degree=3, C=C).fit(X, Y)

lin_svc = svm.LinearSVC(C=C).fit(X, Y)

# create a mesh to plot in
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

# title for the plots
titles = ['SVC with linear kernel',
          'SVC with RBF kernel',
          'SVC with polynomial (degree 3) kernel',
          'LinearSVC (linear kernel)']


for i, clf in enumerate((svc, rbf_svc, poly_svc, lin_svc)):
    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, m_max]x[y_min, y_max].
    plt.subplot(2, 2, i + 1)
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
    plt.axis('off')

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=Y, cmap=plt.cm.Paired)

    plt.title(titles[i])
```

For more, see: [Support Vector Machines (in scikit-learn)](http://scikit-learn.org/stable/modules/svm.html), and [the code for the plotting](http://scikit-learn.org/stable/auto_examples/svm/plot_iris.html)
