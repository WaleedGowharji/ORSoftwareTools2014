# Classification and Regression Trees (CART)
In this section, we'll learn how to work with scikit-learn for CART and random forests

## Preprocessing

```python
import numpy as np
import pandas as pd
```

```python
# See IntroPython2.ipynb
titanic_passengers = pd.read_csv("TitanicPassengers.csv").fillna(0)
titanic_passengers
indices = np.arange(len(titanic_passengers))
np.random.shuffle(indices)
data_train, data_test = titanic_passengers.ix[indices[:5*756/6],:], titanic_passengers.ix[indices[5*756/6:],:]
```

## Decision Trees

```python
from sklearn import tree
titanic_CART = tree.DecisionTreeClassifier()
```

```python
data = {'class': data_train.Class.apply(lambda x: int(x[0])),
        'sex': data_train.Sex.apply(lambda x: 0 if x=='female' else 1),
        'age':data_train.Age}
data = pd.DataFrame(data)
data
```

```python
titanic_CART.fit(data,data_train.Survived)
```

```python
titanic_CART.predict(pd.DataFrame({'class': data_test.Class.apply(lambda x: int(x[0])),
                                   'sex': data_test.Sex.apply(lambda x: 0 if x=='female' else 1),
                                   'age':data_test.Age}))
```
